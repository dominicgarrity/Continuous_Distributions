```python
#import sys
#!{sys.executable} -m pip install -r requirements.txt
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.stats import uniform
from scipy.stats import expon
from scipy.stats import norm
```

```python
## Simulate and plot uniform distribution

# Set up figure and axes for plotting
fig, axs = plt.subplots(3, 3, figsize = (30,30))
# Create counter to track number of iterations and calculate row and col values
k = 0
for i in range(10, 19):
    # Calculate row number
    row = k // 3
    # Calculate column number
    col = k % 3
    
    # Define parameters
    start = 0
    width = 1
    # Calculate number of random variates in experiment
    n = 2 ** i
    
    # Create distribution
    data_uniform = uniform.rvs(size = n, loc = start, scale = width)
    # Create subplot
    ax = axs[row, col]
    ax.hist(data_uniform, bins = 50, color = "lightblue", density = True)
    title = 'Uniform Distribution with ' + str(n) + " Random Variates"
    ax.set_title(title, size = 20)
    
    # Add probability density function
    rv_uniform = uniform(loc = start, scale = width)
    x = np.linspace((start - 0.5), (width + 0.5))
    h = ax.plot(x, rv_uniform.pdf(x), lw = 2, color = "darkred")
    
    # Increment counter
    k += 1
    
# Add labels to axes
for ax in axs.flat:
    ax.set_xlabel('Value', fontsize = 20)
    ax.set_ylabel('Frequency', fontsize = 20)

# Hide X labels and tick labels for top plots and Y ticks for right plots
for ax in axs.flat:
    ax.label_outer()

# Add title to entire figure
fig.suptitle("Uniform Distribution with Varying Numbers of Random Variates", fontweight = "bold", size = 40)

plt.show()
fig.savefig('Plots/Uniform_Distribution.png')
```
